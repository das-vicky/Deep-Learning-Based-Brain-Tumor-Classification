```python
import pandas as pd
import os
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import KFold
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, f1_score, roc_auc_score, roc_curve, auc
from sklearn.utils import resample
from tensorflow.keras import layers, Model, optimizers
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.models import load_model

# --- 1. SETUP & IMPORTS ---
# Force channels_last to ensure consistency
tf.keras.backend.set_image_data_format('channels_last')

# Import EfficientNet
from tensorflow.keras.applications import EfficientNetB0
from tensorflow.keras.applications.efficientnet import preprocess_input

# --- 2. CONFIGURATION ---
TRAIN_DIR = 'archive/Training' 
TEST_DIR = 'archive/Testing'
IMG_SIZE = 224
BATCH_SIZE = 32
N_FOLDS = 5
CLASSES = ['glioma', 'meningioma', 'pituitary', 'notumor']
NUM_CLASSES = 4

# --- 3. PREPARE TRAINING DATA ---
filepaths = []
labels = []

if not os.path.exists(TRAIN_DIR):
    print(f"WARNING: Training directory '{TRAIN_DIR}' not found.")
else:
    classes_list = sorted(os.listdir(TRAIN_DIR))
    for class_name in classes_list:
        class_dir = os.path.join(TRAIN_DIR, class_name)
        if os.path.isdir(class_dir):
            for filename in os.listdir(class_dir):
                filepaths.append(os.path.join(class_dir, filename))
                labels.append(class_name)

df = pd.DataFrame({'filename': filepaths, 'class': labels})
print(f"Found {len(df)} training images.")

# --- 4. CUSTOM ATTENTION LAYER ---
class ChannelAttention(layers.Layer):
    def __init__(self, **kwargs):
        super(ChannelAttention, self).__init__(**kwargs)

    def build(self, input_shape):
        self.channels = input_shape[-1]
        self.w = self.add_weight(name='attention_weights',
                                 shape=(self.channels,),
                                 initializer='ones', 
                                 trainable=True)
        super(ChannelAttention, self).build(input_shape)

    def call(self, inputs):
        alpha = tf.nn.softmax(self.w)
        return inputs * alpha

    def get_config(self):
        return super(ChannelAttention, self).get_config()

# --- 5. ROBUST MODEL BUILDER (MANUAL WEIGHT LOAD) ---
def build_robust_efficientnet_model():
    # Clear session to start fresh
    tf.keras.backend.clear_session()
    
    # 1. Define Input
    inputs = layers.Input(shape=(IMG_SIZE, IMG_SIZE, 3))
    
    # 2. Build Architecture with weights=None
    # This forces Keras to build exactly what we ask for (3 channels), 
    # preventing the "expect 1 channel" error.
    base_model = EfficientNetB0(
        weights=None,  # <--- CRITICAL: Build structure first
        include_top=False, 
        input_tensor=inputs
    )
    
    # 3. Manually Load ImageNet Weights
    # This bypasses the constructor check that was failing
    weights_url = "https://storage.googleapis.com/keras-applications/efficientnetb0_notop.h5"
    weights_path = tf.keras.utils.get_file("efficientnetb0_notop.h5", weights_url)
    
    print("Loading weights manually...")
    try:
        base_model.load_weights(weights_path)
    except Exception as e:
        print(f"Standard load warning: {e}")
        print("Attempting loose load (by_name=True)...")
        base_model.load_weights(weights_path, by_name=True, skip_mismatch=True)
        
    base_model.trainable = True 

    # 4. Custom Head
    x = base_model.output
    x = layers.GlobalAveragePooling2D()(x)
    x = ChannelAttention(name='channel_attention')(x)
    
    x = layers.BatchNormalization()(x) 
    x = layers.Dropout(0.4)(x)
    
    x = layers.Dense(256, activation='relu')(x)
    x = layers.BatchNormalization()(x)
    x = layers.Dropout(0.3)(x)

    outputs = layers.Dense(NUM_CLASSES, activation='softmax')(x)

    model = Model(inputs=inputs, outputs=outputs, name='EfficientNetB0_Robust')
    return model

# --- 6. K-FOLD TRAINING LOOP ---
kf = KFold(n_splits=N_FOLDS, shuffle=True, random_state=42)

# Augmentation (Explicit RGB)
train_datagen_args = dict(
    preprocessing_function=preprocess_input,
    rotation_range=40, 
    width_shift_range=0.2, 
    height_shift_range=0.2, 
    zoom_range=0.2, 
    horizontal_flip=True, 
    fill_mode='nearest'
)
val_datagen_args = dict(preprocessing_function=preprocess_input)

acc_per_fold = []
fold_histories = []
fold_no = 1

if len(df) > 0:
    for train_index, val_index in kf.split(df):
        print(f"\n--- Training Fold {fold_no} / {N_FOLDS} ---")
        train_df = df.iloc[train_index]
        val_df = df.iloc[val_index]
        
        train_datagen = ImageDataGenerator(**train_datagen_args)
        val_datagen = ImageDataGenerator(**val_datagen_args)
        
        train_generator = train_datagen.flow_from_dataframe(
            train_df, x_col='filename', y_col='class',
            target_size=(IMG_SIZE, IMG_SIZE),
            batch_size=BATCH_SIZE,
            class_mode='categorical',
            color_mode='rgb', # Explicitly RGB
            shuffle=True
        )
        
        val_generator = val_datagen.flow_from_dataframe(
            val_df, x_col='filename', y_col='class',
            target_size=(IMG_SIZE, IMG_SIZE),
            batch_size=BATCH_SIZE,
            class_mode='categorical',
            color_mode='rgb', # Explicitly RGB
            shuffle=False
        )
        
        # Build Model
        model = build_robust_efficientnet_model()
        
        model.compile(optimizer=optimizers.Adam(learning_rate=1e-5), 
                      loss='categorical_crossentropy', metrics=['accuracy'])
        
        early_stop = EarlyStopping(monitor='val_loss', patience=8, restore_best_weights=True)
        reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=4, min_lr=1e-7)
        
        history = model.fit(
            train_generator,
            epochs=50, 
            validation_data=val_generator,
            callbacks=[early_stop, reduce_lr],
            verbose=1
        )
        fold_histories.append(history)
        
        scores = model.evaluate(val_generator, verbose=0)
        print(f"Fold {fold_no} Accuracy: {scores[1]*100:.2f}%")
        acc_per_fold.append(scores[1] * 100)
        
        # Save model
        model.save(f"EfficientNetB0_Fold{fold_no}_Best.h5")
        fold_no += 1

    print(f"\nAverage Validation Accuracy: {np.mean(acc_per_fold):.2f}%")

# --- 7. INDEPENDENT TEST & EVALUATION ---
print("\n--- Independent Test Evaluation ---")

test_filepaths = []
test_labels = []

if os.path.exists(TEST_DIR):
    test_classes = sorted(os.listdir(TEST_DIR))
    for class_name in test_classes:
        class_dir = os.path.join(TEST_DIR, class_name)
        if os.path.isdir(class_dir):
            for filename in os.listdir(class_dir):
                test_filepaths.append(os.path.join(class_dir, filename))
                test_labels.append(class_name)

    test_df = pd.DataFrame({'filename': test_filepaths, 'class': test_labels})
    print(f"Found {len(test_df)} Test images.")

    test_datagen = ImageDataGenerator(preprocessing_function=preprocess_input)
    test_gen = test_datagen.flow_from_dataframe(
        test_df, x_col='filename', y_col='class',
        target_size=(IMG_SIZE, IMG_SIZE),
        batch_size=BATCH_SIZE,
        class_mode='categorical',
        color_mode='rgb',
        shuffle=False
    )

    # Ensemble Prediction
    ensemble_sum = np.zeros((len(test_df), NUM_CLASSES))
    
    for fold_idx in range(1, N_FOLDS + 1):
        model_name = f"EfficientNetB0_Fold{fold_idx}_Best.h5"
        if os.path.exists(model_name):
            print(f"Predicting with {model_name}...")
            # Load with Custom Object
            fold_model = load_model(model_name, custom_objects={'ChannelAttention': ChannelAttention})
            ensemble_sum += fold_model.predict(test_gen, verbose=0)
    
    test_predictions = ensemble_sum / N_FOLDS
    y_pred = np.argmax(test_predictions, axis=1)
    y_true = test_gen.classes
    
    # --- OUTPUTS ---
    
    # 1. Classification Report
    print("\nClassification Report:")
    print(classification_report(y_true, y_pred, target_names=CLASSES))
    
    # 2. Confusion Matrix
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=CLASSES, yticklabels=CLASSES)
    plt.title('EfficientNetB0 Ensemble Confusion Matrix')
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    plt.show()
    
    # 3. Training History (Last Fold)
    if len(fold_histories) > 0:
        lh = fold_histories[-1]
        plt.figure(figsize=(12, 5))
        plt.subplot(1, 2, 1)
        plt.plot(lh.history['accuracy'], label='Train')
        plt.plot(lh.history['val_accuracy'], label='Val')
        plt.title('Accuracy (Last Fold)')
        plt.legend()
        plt.subplot(1, 2, 2)
        plt.plot(lh.history['loss'], label='Train')
        plt.plot(lh.history['val_loss'], label='Val')
        plt.title('Loss (Last Fold)')
        plt.legend()
        plt.show()

    # 4. ROC-AUC
    y_true_oh = tf.keras.utils.to_categorical(y_true, num_classes=NUM_CLASSES)
    fpr, tpr, roc_auc = {}, {}, {}
    plt.figure(figsize=(10, 8))
    for i in range(NUM_CLASSES):
        fpr[i], tpr[i], _ = roc_curve(y_true_oh[:, i], test_predictions[:, i])
        roc_auc[i] = auc(fpr[i], tpr[i])
        plt.plot(fpr[i], tpr[i], label=f'{CLASSES[i]} (AUC = {roc_auc[i]:.2f})')
    plt.plot([0, 1], [0, 1], 'k--')
    plt.title('ROC-AUC Curve')
    plt.legend()
    plt.show()
    
    # 5. Save Wilcoxon Data
    results_df = pd.DataFrame({
        'filename': test_gen.filenames,
        'y_true': y_true,
        'y_pred': y_pred,
        'prob_true_class': [test_predictions[i, label] for i, label in enumerate(y_true)],
        'is_correct': (y_true == y_pred).astype(int)
    })
    results_df.to_csv('EfficientNetB0_Wilcoxon_Data.csv', index=False)
    print("Saved 'EfficientNetB0_Wilcoxon_Data.csv'.")
    
    # 6. Bootstrapping (95% CI)
    print("\nCalculating 95% CI (Bootstrapping)...")
    stats = {'accuracy': [], 'f1': [], 'auc': []}
    n_iterations = 1000
    y_true_np = np.array(y_true)
    y_pred_np = np.array(y_pred)
    
    for i in range(n_iterations):
        indices = resample(np.arange(len(y_true)), replace=True)
        y_true_b = y_true_np[indices]
        y_pred_b = y_pred_np[indices]
        y_prob_b = test_predictions[indices]
        
        stats['accuracy'].append(accuracy_score(y_true_b, y_pred_b))
        stats['f1'].append(f1_score(y_true_b, y_pred_b, average='weighted'))
        try:
            y_true_b_onehot = tf.keras.utils.to_categorical(y_true_b, num_classes=NUM_CLASSES)
            if y_true_b_onehot.shape[1] == NUM_CLASSES:
                 stats['auc'].append(roc_auc_score(y_true_b_onehot, y_prob_b, multi_class='ovr', average='weighted'))
        except:
            pass

    for key, values in stats.items():
        if values:
            lower = np.percentile(values, 2.5)
            upper = np.percentile(values, 97.5)
            print(f"{key.upper()}: {np.mean(values):.4f} (95% CI: {lower:.4f} - {upper:.4f})")
```
