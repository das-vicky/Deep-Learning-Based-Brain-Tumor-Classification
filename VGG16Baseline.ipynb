```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, Model
from tensorflow.keras.applications import VGG16
from tensorflow.keras.applications.vgg16 import preprocess_input
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from sklearn.metrics import f1_score, roc_auc_score, accuracy_score
from sklearn.preprocessing import LabelBinarizer
import scipy.stats as stats
import numpy as np
import pandas as pd
import os
from sklearn.model_selection import KFold
from sklearn.utils.class_weight import compute_class_weight
import matplotlib.pyplot as plt

# --- INITIALIZE METRIC LISTS ---
fold_acc_scores = []
fold_f1_scores = []
fold_auc_scores = []

# --- 1. Parameters & Constants ---
IMG_SIZE = 224
BATCH_SIZE = 32
CLASSES = ['glioma', 'meningioma', 'pituitary', 'notumor']
NUM_CLASSES = 4
TRAIN_DIR = 'archive/Training'
TEST_DIR = 'archive/Testing'
N_FOLDS = 5
EPOCHS = 50  # Reduced for CV efficiency, adjust as needed

# --- 2. Data Preparation: Create a DataFrame ---
# We scan the directory manually to create a list of all files.
# This allows us to use KFold on the file lists.
file_paths = []
labels = []

for class_name in CLASSES:
    class_dir = os.path.join(TRAIN_DIR, class_name)
    if os.path.isdir(class_dir):
        for fname in os.listdir(class_dir):
            if fname.lower().endswith(('.png', '.jpg', '.jpeg')):
                file_paths.append(os.path.join(class_dir, fname))
                labels.append(class_name)

# Create a master DataFrame
full_train_df = pd.DataFrame({
    'filename': file_paths,
    'class': labels
})

# Shuffle the dataframe
full_train_df = full_train_df.sample(frac=1, random_state=42).reset_index(drop=True)

print(f"Total training images found: {len(full_train_df)}")

# --- 3. Model Building Function ---
def build_baseline_vgg_model(input_shape=(IMG_SIZE, IMG_SIZE, 3), num_classes=4):
    base_model = VGG16(weights='imagenet', include_top=False, input_shape=input_shape)
    base_model.trainable = False  # Freeze VGG16
    
    inputs = base_model.input
    x = base_model.output
    x = layers.GlobalAveragePooling2D(name='global_avg_pool')(x)
    outputs = layers.Dense(num_classes, activation='softmax')(x)

    model = Model(inputs=inputs, outputs=outputs, name='VGG16_Baseline')
    
    optimizer = keras.optimizers.Adam(learning_rate=1e-5)
    model.compile(
        optimizer=optimizer,
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )
    return model

# --- 4. K-Fold Cross Validation Loop ---
kf = KFold(n_splits=N_FOLDS, shuffle=True, random_state=42)

fold_no = 1
acc_per_fold = []
loss_per_fold = []
histories = []

# Initialize the DataGenerator (Augmentation for train, simple pre-proc for val)
train_datagen_core = ImageDataGenerator(
    preprocessing_function=preprocess_input,
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)

val_datagen_core = ImageDataGenerator(
    preprocessing_function=preprocess_input
)

for train_index, val_index in kf.split(full_train_df):
    print(f"\nTraining for Fold {fold_no} ...")
    
    # 1. Split DataFrames for this fold
    train_df = full_train_df.iloc[train_index]
    val_df = full_train_df.iloc[val_index]
    
    # 2. Create Generators using flow_from_dataframe
    train_generator = train_datagen_core.flow_from_dataframe(
        dataframe=train_df,
        x_col='filename',
        y_col='class',
        target_size=(IMG_SIZE, IMG_SIZE),
        batch_size=BATCH_SIZE,
        class_mode='categorical',
        shuffle=True
    )
    
    val_generator = val_datagen_core.flow_from_dataframe(
        dataframe=val_df,
        x_col='filename',
        y_col='class',
        target_size=(IMG_SIZE, IMG_SIZE),
        batch_size=BATCH_SIZE,
        class_mode='categorical',
        shuffle=False
    )
    
    # 3. Calculate Class Weights (Specific to this fold's training set)
    y_train_indices = train_generator.classes
    class_labels_unique = np.unique(y_train_indices)
    class_weights = compute_class_weight(
        class_weight='balanced',
        classes=class_labels_unique,
        y=y_train_indices
    )
    class_weights_dict = dict(zip(class_labels_unique, class_weights))

    # 4. Clear Session & Build New Model
    tf.keras.backend.clear_session()
    model = build_baseline_vgg_model()
    
    # 5. Callbacks
    # We save weights specific to the fold
    checkpoint = tf.keras.callbacks.ModelCheckpoint(
        filepath=f'VGG16baselinemodel_fold_{fold_no}.keras',  # Saves as VGG16baselinemodel_fold_1.keras, VGG16baselinemodel_fold_2.keras, etc.
        monitor='val_loss',
        save_best_only=True,
        mode='min',
        verbose=1
    )
    callbacks = [
        EarlyStopping(monitor='val_loss', patience=8, restore_best_weights=True, verbose=1),
        ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=4, min_lr=1e-7, verbose=1)
    ]

    # 6. Train
    history = model.fit(
        train_generator,
        epochs=EPOCHS,
        validation_data=val_generator,
        callbacks=callbacks,
        class_weight=class_weights_dict,
        verbose=1
    )
    
    histories.append(history)
    
    # 7. Evaluate on Validation Split of this fold
    print(f"Score for Fold {fold_no}:")
    scores = model.evaluate(val_generator, verbose=0)
    print(f"{model.metrics_names[0]} of {scores[0]}; {model.metrics_names[1]} of {scores[1]*100}%")
    
    acc_per_fold.append(scores[1] * 100)
    loss_per_fold.append(scores[0])
    # --- NEW CODE: Calculate Detailed Metrics for CI ---
    print("Calculating detailed metrics for this fold...")
    
    # 1. Reset generator to ensure order
    val_generator.reset()
    
    # 2. Get predictions
    y_pred_probs = model.predict(val_generator, verbose=0)
    y_pred_classes = np.argmax(y_pred_probs, axis=1)
    y_true = val_generator.classes
    
    # 3. Calculate & Store Metrics
    # Accuracy
    acc = accuracy_score(y_true, y_pred_classes)
    fold_acc_scores.append(acc)
    
    # F1 Score
    f1 = f1_score(y_true, y_pred_classes, average='weighted')
    fold_f1_scores.append(f1)
    
    # AUC Score
    lb = LabelBinarizer()
    lb.fit(y_true)
    y_true_bin = lb.transform(y_true)
    try:
        auc_val = roc_auc_score(y_true_bin, y_pred_probs, multi_class='ovr', average='weighted')
    except:
        auc_val = 0.0
    fold_auc_scores.append(auc_val)
    fold_no += 1

# --- 5. CV Results Summary ---
print("\n" + "-"*30)
print("Average Scores for all folds:")
print(f"> Accuracy: {np.mean(acc_per_fold):.2f} (+- {np.std(acc_per_fold):.2f})")
print(f"> Loss: {np.mean(loss_per_fold):.4f}")
print("-"*30)

# --- NEW CODE: Save Fold Metrics for Wilcoxon Test ---
results_df = pd.DataFrame({
    'Fold': range(1, N_FOLDS + 1),
    'Accuracy': acc_per_fold,
    'Loss': loss_per_fold
})

# Save to CSV
results_df.to_csv('VGG16_Baseline_Fold_Scores.csv', index=False)
print("Fold scores saved to 'VGG16_Baseline_Fold_Scores.csv'")
# --- 6. Final Visualization (Average Accuracy/Loss) ---
# Create an average history plot
plt.figure(figsize=(12, 4))

# Plot Accuracy
plt.subplot(1, 2, 1)
for i, h in enumerate(histories):
    plt.plot(h.history['accuracy'], label=f'Fold {i+1} Train', alpha=0.3)
    plt.plot(h.history['val_accuracy'], label=f'Fold {i+1} Val')
plt.title('Model Accuracy across Folds')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(loc='lower right')

# Plot Loss
plt.subplot(1, 2, 2)
for i, h in enumerate(histories):
    plt.plot(h.history['loss'], label=f'Fold {i+1} Train', alpha=0.3)
    plt.plot(h.history['val_loss'], label=f'Fold {i+1} Val')
plt.title('Model Loss across Folds')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend()

plt.show()

# # --- 7. Final Test Evaluation (Optional) ---
# # Usually in CV, you select the best hyperparameters, retrain on ALL data, and then test.
# # Or you can test the model from the last fold (or best fold) just to see performance on unseen data.
# print("\nEvaluating Last Fold Model on Unseen Test Set:")
# test_datagen = ImageDataGenerator(preprocessing_function=preprocess_input)
# test_generator = test_datagen.flow_from_directory(
#     TEST_DIR,
#     target_size=(IMG_SIZE, IMG_SIZE),
#     batch_size=BATCH_SIZE,
#     class_mode='categorical',
#     classes=CLASSES,
#     shuffle=False
# )
# model.evaluate(test_generator)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc
from sklearn.preprocessing import label_binarize
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

# Helper function to plot everything
def evaluate_and_plot(model, generator, dataset_name, class_names):
    print(f"\n--- Evaluating on {dataset_name} Set ---")
    
    # 1. Get Predictions
    # Important: Reset generator to start to ensure order matches
    generator.reset() 
    y_pred_probs = model.predict(generator, verbose=0)
    y_pred_classes = np.argmax(y_pred_probs, axis=1)
    y_true = generator.classes
    
    # 2. Classification Report
    print(f"\nClassification Report ({dataset_name}):")
    print(classification_report(y_true, y_pred_classes, target_names=class_names))
    
    # 3. Confusion Matrix
    cm = confusion_matrix(y_true, y_pred_classes)
    plt.figure(figsize=(14, 6))
    
    # Plot Confusion Matrix
    plt.subplot(1, 2, 1)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
                xticklabels=class_names, yticklabels=class_names)
    plt.title(f'Confusion Matrix - {dataset_name}')
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    
    # 4. ROC-AUC Curve (Multiclass)
    # Binarize the true labels for ROC calculation
    y_true_bin = label_binarize(y_true, classes=range(len(class_names)))
    n_classes = len(class_names)
    
    plt.subplot(1, 2, 2)
    
    # Calculate ROC for each class
    for i in range(n_classes):
        fpr, tpr, _ = roc_curve(y_true_bin[:, i], y_pred_probs[:, i])
        roc_auc = auc(fpr, tpr)
        plt.plot(fpr, tpr, lw=2, label=f'{class_names[i]} (AUC = {roc_auc:.2f})')
    
    plt.plot([0, 1], [0, 1], 'k--', lw=2) # Random guess line
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(f'ROC-AUC Curve - {dataset_name}')
    plt.legend(loc="lower right")
    
    plt.tight_layout()
    plt.show()

# Get class names from the generator (e.g., ['glioma', 'meningioma', ...])
class_labels = list(test_generator.class_indices.keys())
```

```python
# --- NEW CODE: Save Prediction Details for McNemar's Test ---
import pandas as pd
import numpy as np

# 1. Ensure Generator is Reset and Unshuffled (CRITICAL for matching rows)
test_datagen = ImageDataGenerator(preprocessing_function=preprocess_input)
test_generator = test_datagen.flow_from_directory(
    TEST_DIR,
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    classes=CLASSES, 
    shuffle=False  # MUST BE FALSE
)

# 2. Get Predictions
print("Generating predictions for Statistical Tests...")
pred_probs = model.predict(test_generator, verbose=1)
pred_classes = np.argmax(pred_probs, axis=1)

# 3. Get True Labels
true_classes = test_generator.classes
filenames = test_generator.filenames

# 4. Extract Probability of the TRUE class (Needed for Wilcoxon)
# We look at the row [i] and select the column [true_label]
prob_true_class = [pred_probs[i, true_classes[i]] for i in range(len(pred_probs))]

# 5. Create DataFrame matching the Optimized model's format
baseline_df = pd.DataFrame({
    'filename': filenames,
    'y_true': true_classes,
    'y_pred': pred_classes,
    'prob_true_class': prob_true_class, # <--- NEW COLUMN ADDED
    'is_correct': (true_classes == pred_classes).astype(int)
})

# 6. Save
baseline_df.to_csv('VGG16_Baseline_Stats_Data.csv', index=False)
print("SUCCESS: Saved 'VGG16_Baseline_Stats_Data.csv' with probabilities.")
```

```python
# --- NEW CODE: Print 95% Confidence Intervals ---
def print_confidence_interval(metric_name, scores):
    mean_score = np.mean(scores)
    std_dev = np.std(scores, ddof=1)
    n = len(scores)
    
    confidence = 0.95
    t_crit = stats.t.ppf((1 + confidence) / 2, df=n-1)
    margin_of_error = t_crit * (std_dev / np.sqrt(n))
    
    lower = mean_score - margin_of_error
    upper = mean_score + margin_of_error
    
    print(f"--- {metric_name} ---")
    print(f"Mean: {mean_score:.4f}")
    print(f"95% CI: [{lower:.4f}, {upper:.4f}]")

print("\n=== Cross-Validation 95% Confidence Intervals ===")
print_confidence_interval("Accuracy", fold_acc_scores)
print_confidence_interval("F1 Score", fold_f1_scores)
print_confidence_interval("AUC", fold_auc_scores)
```
